```python
# Tum ogrencilerin bilgilerini tek bir dataframe icinde su sutunlardan olusacak sekilde birlestiriniz; > 
#" Sinif_Adi, Ogrenci Adi, Dogru Sayisi, Yanlis Sayisi, Bos Sayisi" .

# Sinava giren ogrenci sayisini hesaplayiniz.
# Siniflarin ayri ayri Dogru yanlis bos ortalamalarini hesaplayiniz.
# Her sinifin ayri ayri en basarili ogrencisini bulunuz.
# Tum sinavin en basarili ilk 3 ogrencisini bulunuz.
# Siniflarin standart sapmalarini hesaplayiniz. (Dogru sayilari uzerinden)
# Sinavin dogru yanlis ve bos ortalamalarini hesaplayiniz.
```

```python
import numpy as np
import pandas as pd
```

```python
liste=[]
sayac=0
data=data=pd.read_excel("py_mind.xlsx",sheet_name=0)
liste+=[{"Sinif_Adi":"py_mind", 
                        "Ogrenci Adi":data.columns[0],
                        "Dogru Sayisi":data.iloc[20][1], 
                        "Yanlis Sayisi":data.iloc[21][1],
                        "Bos Sayisi":data.iloc[22][1]}]
```

```python

for i in range(1,10):
       
    data=pd.read_excel("py_mind.xlsx",sheet_name=i)
    data=data.dropna()
    if data.size== 0:
        sayac+=1
        liste+=[{"Sinif_Adi":"py_mind", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":0, 
                 "Yanlis Sayisi":0,
                 "Bos Sayisi":0}]
    else:
        liste+=[{"Sinif_Adi":"py_mind", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":data.iloc[0][1], 
                 "Yanlis Sayisi":data.iloc[1][1],
                 "Bos Sayisi":data.iloc[2][1]}]

```

```python
for i in range(0,10):
       
    data=pd.read_excel("py_sense.xlsx",sheet_name=i)
    data=data.dropna()
    if data.size== 0:
        sayac+=1
        liste+=[{"Sinif_Adi":"py_sense", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":0, 
                 "Yanlis Sayisi":0,
                 "Bos Sayisi":0}]
    else:
        liste+=[{"Sinif_Adi":"py_sense", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":data.iloc[0][1], 
                 "Yanlis Sayisi":data.iloc[1][1],
                 "Bos Sayisi":data.iloc[2][1]}]
```

```python

for i in range(0,10):
       
    data=pd.read_excel("py_opinion.xlsx",sheet_name=i)
    data=data.dropna()
    if data.size== 0:
        sayac+=1
        liste+=[{"Sinif_Adi":"py_opinion", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":0, 
                 "Yanlis Sayisi":0,
                 "Bos Sayisi":0}]
    else:
        liste+=[{"Sinif_Adi":"py_opinion", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":data.iloc[0][1], 
                 "Yanlis Sayisi":data.iloc[1][1],
                 "Bos Sayisi":data.iloc[2][1]}]

```

```python
for i in range(0,8):
       
    data=pd.read_excel("py_science.xlsx",sheet_name=i)
    data=data.dropna()
    if data.size== 0:
        sayac+=1
        liste+=[{"Sinif_Adi":"py_science", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":0, 
                 "Yanlis Sayisi":0,
                 "Bos Sayisi":0}]
    else:
        liste+=[{"Sinif_Adi":"py_science", 
                 "Ogrenci Adi":data.columns[0],
                 "Dogru Sayisi":data.iloc[0][1], 
                 "Yanlis Sayisi":data.iloc[1][1],
                 "Bos Sayisi":data.iloc[2][1]}]
toplam=len(liste)
```
